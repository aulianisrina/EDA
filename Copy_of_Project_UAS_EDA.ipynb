# Gen Z yang ingin Childfree

Akan dianalisis lebih lanjut mengenai para responden yang memilih untuk childfree

```python
#Membentuk DataSet baru dengan syarat khusus asal negara Pakistan
CF = df.loc[df['Apakah kamu berkeinginan untuk Childfree?'] == 'Ya']
CF
```

```python
#Membentuk DataSet baru sesuai dengan variabel-variabel yang diperlukan
CF1 = CF.reset_index()[['Nama/Inisial','Umur','Faktor apa yang menjadi alasan kamu memilih untuk Childfree?', 'Apakah keluargamu/pasanganmu mendukung keputusanmu memilih Childfree?', 'Apakah dulu kamu pernah berkeinginan untuk memiliki anak?', 'Apakah kamu pernah mendapat kritikan/perlakuan berbeda karena keputusanmu untuk childfree?', 'Bagaimana perasaanmu saat ini setelah memutuskan untuk Childfree?', 'Apakah ada kemungkinan kamu akan merubah keputusanmu untuk childfree di masa depan?']]
CF1
```

```python
#fungsi rekursif untuk melihat distribusi data
sns.set_theme()

def draw_distribution_graph(dataframe, column_name):

    fig, ax = plt.subplots(figsize=(8, 8))
    sns.distplot(dataframe[column_name], color='blue', ax = ax).set_title("Distribution Graph for {}".format(column_name), fontsize=16)
    plt.xlabel(column_name, fontsize=12)
    plt.ylabel('Density', fontsize=12)
    plt.show()
```

```python
draw1 = draw_distribution_graph(CF1, 'Apakah keluargamu/pasanganmu mendukung keputusanmu memilih Childfree?')
```

Di lihat dari grafik diketahui bahwa keputusan mereka untuk childfree masih belum didukung sepenuhnya oleh keluarga/pasangan mereka.

keterangan:

*   Skala 5 menunjukkan "sangat didukung"
*   Skala 1 menunjukkan "sangat ditentang"

```python
draw2 = draw_distribution_graph(CF1, 'Apakah dulu kamu pernah berkeinginan untuk memiliki anak?')
```

Pada grafik diatas, terlihat bahwa banyak dari mereka dulu pernah mempunyai keinginan untuk memiliki anak. Hal ini dikarenakan skala 4 dan 3 paling banyak dipilih oleh responden.

keterangan:

*   Skala 5 menunjukkan "sangat pernah"
*   Skala 1 menunjukkan "tidak pernah"

```python
draw3 = draw_distribution_graph(CF1, 'Apakah kamu pernah mendapat kritikan/perlakuan berbeda karena keputusanmu untuk childfree?')
```

Pada grafik diatas, dapat diketahui bahwa masih terdapat beberapa dari mereka yang mendapat kritikan dan diskriminasi oleh orang lain karena memutuskan untuk childfree. Hal ini dikarenakan skala 4 dan 5 paling banyak dipilih oleh responden.

keterangan:

*   Skala 5 menunjukkan "sangat pernah"
*   Skala 1 menunjukkan "tidak pernah"

```python
draw4 = draw_distribution_graph(CF1, 'Bagaimana perasaanmu saat ini setelah memutuskan untuk Childfree?')
```

Pada grafik diatas, terlihat bahwa sebagian besar dari mereka sudah puas dan senang setelah memutuskan untuk childfree. Hal ini dikarenakan skala 4 dan 5 paling banyak dipilih oleh responden.

keterangan:

*   Skala 5 menunjukkan "sangat senang"
*   Skala 1 menunjukkan "sangat menyesal"

```python
draw5 = draw_distribution_graph(CF1, 'Apakah ada kemungkinan kamu akan merubah keputusanmu untuk childfree di masa depan?')
```

Pada grafik diatas, terlihat bahwa sebagian besar dari mereka masih tidak yakin apakah nantinya mereka akan berubah pikiran atau tidak. Hal ini dikarenakan skala 3 paling banyak dipilih oleh responden. Terdapat pula beberapa yang sudah yakin untuk tidak akan merubah keputusannya untuk tidak memiliki anak.

keterangan:

*   Skala 5 menunjukkan "sangat mungkin"
*   Skala 1 menunjukkan "tidak mungkin"

```python
#Untuk melihat outlier yang ada pada data
fig, ax = plt.subplots(5,2, figsize = (10,10))
ax[0,0].boxplot(CF1['Apakah keluargamu/pasanganmu mendukung keputusanmu memilih Childfree?'], 10)
ax[1,0].boxplot(CF1['Apakah dulu kamu pernah berkeinginan untuk memiliki anak?'], 10)
ax[2,0].boxplot(CF1['Apakah kamu pernah mendapat kritikan/perlakuan berbeda karena keputusanmu untuk childfree?'], 10)
ax[3,0].boxplot(CF1['Bagaimana perasaanmu saat ini setelah memutuskan untuk Childfree?'], 10)
ax[4,0].boxplot(CF1['Apakah ada kemungkinan kamu akan merubah keputusanmu untuk childfree di masa depan?'], 10)


ax[0,0].title.set_text('Apakah keluargamu/pasanganmu mendukung keputusanmu memilih Childfree?')
ax[1,0].title.set_text('Apakah dulu kamu pernah berkeinginan untuk memiliki anak?')
ax[2,0].title.set_text('Apakah kamu pernah mendapat kritikan/perlakuan berbeda karena keputusanmu untuk childfree?')
ax[3,0].title.set_text('Bagaimana perasaanmu saat ini setelah memutuskan untuk Childfree?')
ax[4,0].title.set_text('Apakah ada kemungkinan kamu akan merubah keputusanmu untuk childfree di masa depan?')


plt.tight_layout()

plt.show()
```

Dari grafik diatas diketahui bahwa masih ada outlier dari data yang kami ambil. Namun, nilai dari outlier yang ada masih dalam rentang nilai yang kami tentukan di awal sehingga tidak akan dilakukan tindakan khusus pada outlier.
